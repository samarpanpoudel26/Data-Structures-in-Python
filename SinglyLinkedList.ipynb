```python
#creating a class Node 
class Node:
    def __init__(self,item=None,next=None):
        self.item=item
        self.next=next
#creating a class SLL : Singly Linked List
class sll:
    def __init__(self,start=None):
        self.start=start
    def insert(self,data):
        n=Node(data,self.start)
        self.start=n
    def append(self,data):
        n=Node(data)
        if self.start is None: # if the linked list is empty
            self.start=n
        else: # if the linked list has more than one node
            temp=self.start
            while temp.next is not None:
                temp=temp.next
            temp.next=n
    def insert_after(self,curr,data): # inserting after the node
        temp=self.start
        while temp is not None:
            if temp.item ==curr:
                n=Node(data,temp.next)
                temp.next=n
            temp=temp.next
        return None
    def remove(self,data): # removing the node of the linked list
        if self.start is None: # if the linked list is null
            pass
        elif self.start.item==data: # if the linked list has one node
            self.start=self.start.next
            return
        else:
            temp=self.start # if the linked list has more than one node
            while temp.next and temp.next.item!=data:
                temp=temp.next
            if temp.next:
                temp.next=temp.next.next
    def remove_after(self,data): # removing after the given node of the linked list
        if self.start is None: # if linked list is empty
            pass
        elif self.start.item==data: # if linked list has one node
            self.start.next=self.start.next.next
            return
        else: # if linked list has more than one node
            temp=self.start
            while temp.next and temp.next.item!=data:
                temp=temp.next
            if temp.next:
                temp.next.next=temp.next.next.next
    def print_list(self):
        temp=self.start
        while temp is not None:
            print(temp.item, end=" ")
            temp=temp.next
    def sort(self): # sorting
     if self.start is None:
        return
    
     swapped = True
     while swapped: #to sort recursively
        swapped = False
        temp = self.start

        while temp.next is not None:
            if temp.item > temp.next.item:
                temp.item, temp.next.item = temp.next.item, temp.item
                swapped = True
            temp = temp.next


#creating an object of class
mylist=sll()
mylist.insert(30)
mylist.insert(20)
mylist.insert(10)
mylist.insert(45)
mylist.insert_after(20,19)
#mylist.print_list()
#mylist.remove_after(20)
mylist.sort()
mylist.print_list()
```
