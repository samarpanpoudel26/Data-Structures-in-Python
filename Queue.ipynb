```python
class Node:
    def __init__(self,item=None,next=None):
        self.item=item
        self.next=next

class queue:
    def __init__(self):
        self.front=None
        self.rear=None
    def is_empty(self): # finding if the queue is empty
        return self.front==None
    def enqueue(self,data): # entering the value in the queue
        n=Node(data)
        if self.front is None:
            self.front=n
            self.rear=n
        else:
            self.rear.next=n
            self.rear=n
    def dequeue(self): #removing the value from the queue
        if self.front is None:
            raise IndexError("Empty Queue")
        elif self.front == self.rear:
            self.front=None
            self.rear=None
        else:
            self.front=self.front.next
    def get_front(self): # get the front value of the queue
        if self.front is None:
            raise IndexError("Empty queue")
        else:
            return self.front.item
    def get_rear(self): # get the last(rear) value of the queue
        if self.front is None:
            raise IndexError("Empty Queue")
        else:
            return self.rear.item
    def display(self): # printing the queue
        if self.front is None:
            raise IndexError("Empty Queue")
        else:
            temp=self.front
            while temp is not None:
                print(temp.item, end=" ")
                temp=temp.next

#creating object of a class and performing operations
q1=queue()
q1.enqueue(10)
q1.enqueue(20)
q1.enqueue(30)
q1.dequeue()
q1.dequeue()
q1.display()
```
