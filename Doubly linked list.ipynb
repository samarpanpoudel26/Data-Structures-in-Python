```python
class Node:
    def __init__(self,prev=None,item=None,next=None):
        self.prev=prev
        self.item=item
        self.next=next

class DLL:
    def __init__(self,start=None):
        self.start=start
    def insert(self,data):
        n=Node(None,data,self.start)
        if self.start is not None: 
            self.start.prev=n
        self.start=n
    def append(self,data):
        temp=self.start
        if temp!=None:
          while temp.next is not None:
            temp=temp.next
        n=Node(temp,data,None)
        if temp==None: # if the linked list is empty
            self.start=n
        else:
            temp.next=n
    def insert_after(self,curr,data): # inserting after the given value
        temp=self.start #assisgning the first node
        while temp is not None: # traversing to find the last node
            if temp.item==curr:
                n=Node(temp,data,temp.next)
                temp.next=n
            temp=temp.next
        return None
    def print(self):
        temp=self.start
        while temp is not None:
            print(temp.item,end=" ")
            temp=temp.next
    def sort(self): #sorting continuously
        if self.start is None: #if the linked list is empty
            return
        swapped=True
        while swapped:
            swapped=False
            temp=self.start
            while temp.next is not None: 
                if temp.item > temp.next.item:
                    temp.item,temp.next.item=temp.next.item,temp.item
                    swapped=True
                temp=temp.next
    def delete(self,value):
        if self.start is None: # if the linked list is empty
            return
        elif self.start.item==value: # if the linked list has only one node
            self.start=self.start.next
            if self.start:
                self.start.prev=None
        else:
            temp=self.start # if linked list has multiple node
            while temp.next and temp.next.item!=value:
                temp=temp.next
            if temp.next:
                delete_node=temp.next
                temp.next=delete_node.next
                if delete_node.next:
                    delete_node.next.prev=temp
    def delete_after(self,value):
        if self.start is None: # if the linked list is empty
            return
        elif self.start.item==value: # if the linked list has only one node
            self.start.next=self.start.next.next
            if self.start.next:
                self.start.next.prev=None
        else: # if the linked list has multiple nodes
            temp=self.start
            while temp.next and temp.next.item!=value:
                temp=temp.next
            if temp.next:
                delete_node=temp.next.next
                temp.next.next=delete_node.next
                if delete_node.next:
                    delete_node.next.prev=temp
    


#Creating object of a class and performing operations

lst=DLL()
lst.insert(10)
#lst.insert(15)
lst.append(20)
lst.insert_after(10,11)
lst.insert_after(20,15)
#lst.delete(11)
lst.delete_after(20)
lst.print()
        

            
```
