```python
class HashTable:
    def __init__(self,size):
        self.size=size
        self.table=[None]*self.size
        #self.data=[None]* self.size
    
    def hash_function(self,key): # hash function to find index
        return key % self.size

    def insert(self,key):
        index=self.hash_function(key) 
        start_index=index # storing the starting index to compare later
        while self.table[index] is not None and self.table[index]!="DELETED":
            if self.table[index]==key:
                print(f"{key} already exists in index {index}")
                return
            index=(index+1)% self.size #increasing the indexing size by 1 to compare
            if index==start_index:
                print("hash table is full")
                return
        self.table[index]=key
        print(f"{key} successfully inserted at index {index}")

    def search(self,key):
        index=self.hash_function(key)
        start_index=index # storing the starting index to compare later

        while self.table[index] is not None:
            if self.table[index]==key:
                print(f"{key} is found at index {index}")
                return
            index=(index+1)%self.size #increasing the indexing size by 1 to compare
            if index==start_index:
                break
        print("key not found")
        return None

    def delete(self,key):
        index=self.hash_function(key)
        start_index=index # storing the starting index to compare later

        while self.table[index] is not None:
            if self.table[index]==key:
                self.table[index]="DELETED"
                print(f"key {key} is successfully deleted at index {index}")
                return
            index=(index+1)%self.size #increasing the indexing size by 1 to compare
            if index==start_index:
                break
        print("key cannot be found")

    def display(self):
        for i in range(self.size):
            print(f"{i}:{self.table[i]}")
```

```python
# Example usage
if __name__ == "__main__":
    numbers = [99, 9, 93, 2, 3, 4, 5, 55, 66, 7]
    h = HashTable(10)

    for num in numbers:
        h.insert(num)

    h.display()

    print("\nSearching for 93 and 55:")
    h.search(93)
    h.search(55)

    print("\nDeleting 55 and 100 (not present):")
    h.delete(55)
    h.delete(100)

    print("\nAfter Deletion:")
    h.display()
```
