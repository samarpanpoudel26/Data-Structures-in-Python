## Adjacency Matrix

```python
class Matrix:
    def __init__(self,vertices):
        self.vertices=vertices
        self.adjacencymatrix=[[0 for _ in range(vertices)] for _ in range(vertices)]
    def add_edge(self,u,v):
        if self.adjacencymatrix[u][v]==0:
            self.adjacencymatrix[u][v]=1
    def print_matrix(self):
        print("Adjacency Matrix \n")
        for row in self.adjacencymatrix:
            print(row)

g_matrix=Matrix(4)
g_matrix.add_edge(0,1)
g_matrix.add_edge(0,2)
g_matrix.add_edge(1,2)
g_matrix.add_edge(2,2)
g_matrix.add_edge(3,0)
g_matrix.print_matrix()
        
```

## Adjacency List

```python
class Node:
    def __init__(self,vertex,next=None):
        self.vertex=vertex
        self.next=next
class List:
    def __init__(self,vertices):
        self.vertices=vertices
        self.adjacencylist=[None]*vertices
    def add_edge(self,u,v):
        n=Node(v)
        if self.adjacencylist[u] is None:
            self.adjacencylist[u]=n
        else:
            temp=self.adjacencylist[u]
            while temp.next is not None:
                temp=temp.next
            temp.next=n
    def print_graph(self):
     print("\n Adjacency List:")
     for i in range(self.vertices):
      print(f"vertex{i}:",end="")
      temp=self.adjacencylist[i]
      while temp:
        print(f"-> {temp.vertex}",end="")
        temp=temp.next
      print()

g_list=List(4)
g_list.add_edge(0,1)
g_list.add_edge(0,2)
g_list.add_edge(1,2)
g_list.add_edge(2,3)
g_list.add_edge(3,0)
g_list.print_graph()
```
