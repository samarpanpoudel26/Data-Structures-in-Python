```python
class Node:
    def __init__(self,item=None,next=None):
        self.item=item
        self.next=next

class stack:
    def __init__(self):
        self.start=None
        self.item_count=0
    def is_empty(self): #finding if the stack is empty or not
        return self.start == None
    def push(self,data): # inserting the value in the stack
        n=Node(data,self.start)
        self.start=n
    def pop(self): # deleting the value in the stack
        if self.start is not None:
            data=self.start.item
            self.start=self.start.next
            return data
        else:
            raise IndexError("Stack is Empty")
    def peek(self): # returning the first value 
        if self.start is not None:
            return self.start.item
        else:
            raise IndexError("Stack is Empty")
    def display(self): # printing whole stack
        if self.start is not None:
         temp=self.start
         while temp is not None:
            print(temp.item, end=" ")
            temp=temp.next
        else:
            return IndexError("stack is empty")
        
# creating object of a class and performing operations
s1=stack()
s1.push(10)
s1.push(20)
s1.push(30)
print("Top element",s1.peek())
print("Remove first element",s1.pop())
s1.peek()
s1.display()
            
```
