```python
class MinHeap:
    def __init__(self):
        self.heap = []
    def insert(self,value): #inserting value on the heap
        self.heap.append(value)
        self.heapify_up(len(self.heap)-1)
    def heapify_up(self,index): # recursive method to insert value
        parent_index=(index-1)//2
        if index>0 and self.heap[parent_index]>self.heap[index]:
            self.heap[parent_index],self.heap[index]=self.heap[index],self.heap[parent_index]
            self.heapify_up(parent_index)
    def remove(self): # removing the value(top) of the heap
        if not self.heap:
            return None
        if len(self.heap)==1:
            return self.heap.pop()
        root=self.heap[0]
        self.heap[0]=self.heap.pop()
        self.heapify_down(0)
        return root
    def heapify_down(self,index): # recursive method to delete
        smallest=index
        left=2 *index+1
        right=2*index+2
        size=len(self.heap)
        if left<size and self.heap[left]<self.heap[smallest]:
            smallest=left
        if right<size and self.heap[right]<self.heap[smallest]:
            smallest=right
        if smallest!=index:
            self.heap[smallest],self.heap[index]=self.heap[index],self.heap[smallest]
            self.heapify_down(smallest)
    def display(self): # displaying whole heap
        print("MinHeap:", self.heap)


# Example usage
min_heap = MinHeap()
min_heap.insert(10)
min_heap.insert(4)
min_heap.insert(15)
min_heap.insert(2)
min_heap.display()
print("Removed:", min_heap.remove())
min_heap.display()
        
```
