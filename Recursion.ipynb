```python
#Factorial
def factorial(num):
    if num<=1:
        return 1
    else:
        return num * factorial(num-1)
print(factorial(5))
```

```python
#Fibonacci
def fibonacci(num):
    if num==0:
        return 0
    elif num==1:
        return 1
    else:
        return fibonacci(num-1)+fibonacci(num-2)
    
print(fibonacci(5))
```

```python
#checking value in the array
def check_value(arr,num):
    if not arr:
        return False
    if arr[0]==num:
        return True
    else:
        return check_value(arr[1:],num)
    
arr=[1,2,3,4,5]
print(check_value(arr,5))
print(check_value(arr,50))
```

```python
#sum of value in the array
def sum(arr):
    if not arr:
        return False
    else:
        return arr[0]+ sum(arr[1:])
    
arr=[1,2,3,4,5]
print(sum(arr))
```

```python
#Palindrome
def palindrome(s):
    if len(s)<=1:
        return True
    if s[0]!=s[-1]:
        return False
    else:
        return palindrome(s[1:-1])
    
print(palindrome("raar"))
```
